This cell imports the required Python libraries.
Pandas is used for data manipulation and analysis.
Matplotlib is used for data visualization.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

The student marks dataset is loaded from a CSV file into a Pandas DataFrame.
This allows us to easily analyze and manipulate the data using Python.
The dataset is displayed to verify that the data has been loaded correctly.
It contains student names and marks for multiple subjects.


```python
df = pd.read_csv("/content/StudentMarks.csv")
df
```

This cell computes descriptive statistics such as mean, standard deviation,
minimum, maximum, and quartiles for all subjects.
These statistics help in understanding the overall performance and spread
of student marks.



```python
df[["English", "Maths", "Java", "IOT", "Physics"]].describe()  #mean,median,std
```

The mode represents the most frequently occurring marks for each subject.
It helps identify commonly achieved scores among students.


```python
df[["English", "Maths", "Java", "IOT", "Physics"]].mode()  #mode
```

Histograms are used to visualize the distribution of marks for each subject.
They show how frequently different score ranges occur and help identify
patterns such as clustering or skewness.


```python
subjects = ["English", "Maths", "Java", "IOT", "Physics"]

for s in subjects:
    plt.hist(df[s])
    plt.title(f"{s} Marks Distribution")
    plt.xlabel("Marks")
    plt.ylabel("Number of Students")
    plt.show()
```

The boxplot provides a summary of marks distribution across all subjects.
It highlights the median, quartiles, and potential outliers, making it easy
to compare score variations between subjects.


```python
subjects = ["English", "Maths", "Java", "IOT", "Physics"]
plt.boxplot(df[subjects])
plt.xticks(range(1, 6), subjects)
plt.title("Marks Distribution Across Subjects")
plt.show()
```

Insights:
- The mean and median scores across subjects are close, indicating a fairly balanced distribution of marks.

- The standard deviation shows moderate variation, meaning student
  performance is neither too scattered nor too concentrated.

- Histograms reveal that the majority of students scored in the medium to
  high score ranges, with fewer students at the extreme low or high ends.

- The boxplot visualization helps in identifying the spread and symmetry
  of the data. Most subjects do not show extreme outliers, indicating
  consistent student performance.

- Some subjects show slightly higher variability compared to others, which
  may reflect differences in subject difficulty or student understanding.
